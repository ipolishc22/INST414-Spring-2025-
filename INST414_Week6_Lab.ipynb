```python
import json
```

```python
import pandas as pd
```

```python
actor_genre = {}

with open("imdb_movies_2000to2022.prolific.json", "r") as IMDB_file:
    for line in IMDB_file:
        movie = json.loads(line)
        genres = movie['genres']
        actors = movie['actors']

        for actor_id, actor_name in actors:
            if actor_id not in actor_genre:
                actor_genre[actor_id] = {}

            for genre in genres:
                if genre not in actor_genre[actor_id]:
                    actor_genre[actor_id][genre] = 0
                actor_genre[actor_id][genre] += 1
```

```python
index = actor_genre.keys()

rows = [actor_genre[k] for k in index]

df = pd.DataFrame(rows, index=index)
df = df.fillna(0)

df
```

```python
df_norm = df.divide(df.sum(axis=1), axis=0)
df_norm.head(10)
```

```python
from scipy.sparse import lil_matrix
import scipy.spatial.distance
```

```python
target_actor_id = "nm1165110"

target_actor = df_norm.loc[target_actor_id]
distances = scipy.spatial.distance.cdist(df_norm,[target_actor],metric='cosine').flatten()

query_distances = list(zip(df.index,distances))

for similar_actor_id, similar_genre_score in sorted(query_distances,key=lambda x:x[1], reverse=False)[:10]:
    print(similar_actor_id, similar_genre_score)
```

```python
# change for eucledean

target_actor_id = "nm1165110"

target_actor = df_norm.loc[target_actor_id]
distances = scipy.spatial.distance.cdist(df_norm,[target_actor],metric='euclidean').flatten()

query_distances = list(zip(df.index,distances))

for similar_actor_id_similar_genre_score in sorted(query_distances,key=lambda x:x[1], reverse=False)[:10]:
    print(similar_actor_id, similar_genre_score)
```
