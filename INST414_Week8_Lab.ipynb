```python
import pandas as pd
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
actor_genre_df = pd.read_csv('imdb_movies_2000to2022.actorXgenre.csv', index_col='actor_id')
```

```python
inertia_values = []
for k in range(2,31):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(actor_genre_df)
    inertia_values.append(kmeans.inertia_)
```

```python
print(inertia_values)
```

```python
prev = inertia_values[0]
for inertia in inertia_values[1:]:
    print(prev-inertia)
    prev = inertia
```

```python
plt.figure(figsize=(15,6))
plt.plot(range(2,31), inertia_values,marker='o')
plt.title('K-Means Inertia')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia Values')
plt.grid(True)
plt.show()
# based on the graph we can state that the optimal k-value is 10, since the graph decreases at a slower rate after it reaches 10.
```
